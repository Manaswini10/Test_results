```python
# plot test_res_C_1.csv
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('test_res_C_1.csv')
x = data.iloc[:, 0]
y = data.iloc[:, 1]

# Plotting
plt.plot(x, y)
plt.xlim(1, 13)
plt.xlabel('count(increasing CCO)')
plt.ylabel('log(S)')
plt.grid(True)
plt.show()
```
